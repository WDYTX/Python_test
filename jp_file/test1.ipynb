```python
#一个数码管显示
import turtle
def drawline(a):
    turtle.pensize(10)
    turtle.pendown() if a else turtle.penup()
    turtle.fd(40)
    turtle.right(90)
def drawnum(b):
    drawline(True) if b in [2,3,4,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,1,3,4,5,6,7,8,9] else drawline(False)
    drawline(True) if b in [0,2,3,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,2,6,8] else drawline(False)
    turtle.left(90)
    drawline(True) if b in [0,4,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,2,3,5,6,7,8,9] else drawline(False)
    drawline(True) if b in [0,1,2,3,4,7,8,9] else drawline(False)
    turtle.hideturtle()
    turtle.done()
drawnum(9)
```

```python
import turtle
def gap():
    turtle.penup()
    turtle.fd(5)
def drawline(a):
    turtle.pensize(10)
    gap()
    turtle.pendown() if a else turtle.penup()
    turtle.fd(40)
    gap()
    turtle.right(90)
def drawnum(b):
    drawline(True) if b in [2,3,4,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,1,3,4,5,6,7,8,9] else drawline(False)
    drawline(True) if b in [0,2,3,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,2,6,8] else drawline(False)
    turtle.left(90)
    drawline(True) if b in [0,4,5,6,8,9] else drawline(False)
    drawline(True) if b in [0,2,3,5,6,7,8,9] else drawline(False)
    drawline(True) if b in [0,1,2,3,4,7,8,9] else drawline(False)
    turtle.left(180)
    turtle.penup()
    turtle.fd(20)
def shift1(num):#显示数字显示
    for i in num:
        drawnum(eval(i))
def main(x):
    turtle.setup(666,666,None,None)
    turtle.penup()
    turtle.fd(-300)
    shift1(x)
    turtle.hideturtle()
    turtle.done()
main("20230630")
```
