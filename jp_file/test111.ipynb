```python
import turtle
a=['purple','red','yellow','pink']
turtle.setup(650, 350, 200, 200)
turtle.penup()
turtle.fd(-250)
turtle.pendown()
turtle.pensize(25)
turtle.seth(-40)
for i in range(4):
    turtle.pencolor(a[i])
    turtle.circle(40, 80)
    turtle.circle(-40, 80)
    turtle.pencolor('green')
turtle.circle(40, 80/2)
turtle.fd(40)
turtle.circle(16, 180)
turtle.fd(40 * 2/3)
```

```python
import random
random.random()
random.randint(1,2)#产生从1到2的随机数
random.randrange(1,6,1)#产生1-5的随机数并且间距加1
random.randrange(1,6,2)#产生1-5的随机数并且间距加2,如1,3,5
random.getrandbits(16)#产生16位比特数的随机数
random.uniform(10,10.1)#产生10-10.1区间的小数随机数
```

```python
#使用蒙特卡洛计算圆周率
time1=perf_counter()
from time import perf_counter
from random import random
hit=0
num=1000*1000
for i in range (num+1):#随机抛下num次，计算处在圆内的数量hit，让其与num相比得到四分之一的圆面积，
    #再乘以4就是圆的面积，由于半径为1所以圆的面积就为π
    x=random()
    y=random()
    distance=pow(x**2+y**2,0.5)
    if distance<=1.0:
        hit+=1
pai=hit/num*4
print("π的值为：{}".format(pai))
print("使用是时间：{:.3f}".format(perf_counter()-time1))
```
