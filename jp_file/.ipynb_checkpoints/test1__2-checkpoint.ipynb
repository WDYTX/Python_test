```python
import torch
a=torch.arange(4.0,requires_grad=True)
b=a*a
print(b)
print(b.sum())
print(b.sum().backward())
print(a.grad)
a.grad.zero_()
c=a*a
u=c.detach()
print(u)
```

```python
import torch
def synthetic_data(w, b, num_examples):
    X = torch.normal(0, 1, (num_examples, len(w)))
    print(X)
    y = torch.matmul(X, w) + b
    print(y)
    y += torch.normal(0, 0.01, y.shape)
    print(y)
    return X, y.reshape((-1, 1))
print(synthetic_data(torch.tensor([0.1,0.2,0.3,0.4]),torch.tensor([0.2,0.2,0.3,0.3]),4))
```
